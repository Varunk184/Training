```python
#ab
def fibonacci(n):
    a = 0
    b = 1
    if n < 0:
        print("Incorrect input")
    elif n == 0:
        return a
    elif n == 1:
        return b
    else:
        for i in range(2, n+1):
            c = a + b
            a = b
            b = c
        return b
 

 
e = int(input())
print(fibonacci(e))
```

```python
import numpy as np
def min_elements_to_remove(array):
    negatives = 0  
    zeros = 0 

    for num in array:
        if num < 0:
            negatives += 1
        elif num == 0:
            zeros += 1


    if negatives == 0 or negatives % 2 == 0:
        return 0

    if zeros == 0:
        return 1

    return 1

arr = np.array(int(input()))
result = min_elements_to_remove(arr)
print(f"Minimum number of elements to remove: {result}")
```

```python
def min_elements_to_remove(array):
    negatives = 0  # Count of negative numbers
    zeros = 0  # Count of zeros

    for num in array:
        if num < 0:
            negatives += 1
        elif num == 0:
            zeros += 1

    # If there are no negatives or if there are even number of negatives,
    # we don't need to remove any element
    if negatives == 0 or negatives % 2 == 0:
        return 0

    # If there are odd number of negatives and no zeros, we need to remove
    # one negative element
    if zeros == 0:
        return 1

    # If there are odd number of negatives and at least one zero, we need to
    # remove one negative element or one zero element
    return 1

# Test the function
arr = [-2, 3, 0, -4, 5]
result = min_elements_to_remove(arr)
print(f"Minimum number of elements to remove: {result}")
```

```python
def printTransactions(m, k, d, name, owned, prices):
    table = [[(0, []) for _ in range(m + 1)] for _ in range(d + 1)]

    for day in range(1, d + 1):
        for stock in range(k):
            for money in range(1, m + 1):
                if money >= prices[stock][day - 1]:
                    remaining_money = money - prices[stock][day - 1]
                    total_money = table[day - 1][remaining_money][0] + prices[stock][day - 1]
                    if total_money > table[day][money][0]:
                        table[day][money] = (total_money, table[day - 1][remaining_money][1] + [(name[stock], owned[stock])])
    max_money = table[d][m][0]
    transactions = table[d][m][1]
    for transaction in transactions:
        stock_name, stock_owned = transaction
        print(f"Buy {stock_owned} shares of {stock_name}")
    print(f"Remaining Money: ${m - max_money}")
    print(f"Total Money Obtained: ${max_money}")
m = 100
k = 3
d = 5
name = ["Stock A", "Stock B", "Stock C"]
owned = [2, 3, 1]
prices = [[10, 20, 30, 40, 50], [15, 25, 35, 45, 55], [8, 16, 24, 32, 40]]
printTransactions(m, k, d, name, owned, prices)
```
